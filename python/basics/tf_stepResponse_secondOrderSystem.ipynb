# Code to Create Transfer Function of Second order sysetm and Plot Step Response

```python
import numpy as np 
import control as ct 
from control.matlab import *
import matplotlib.pyplot as plt
```

# Define Second Order Transfer functin using num and dem

```python
num=np.array([1,2])
den=np.array([1,2,5])
sys=ct.tf(num,den)
print(sys)
```

# Print Step response Info

```python
a=ct.step_info(sys)
print("Peak : ",a['Peak'])
print(a)
```

# Step Response of above TF and plot on graph   

```python
t,yout=ct.step_response(sys)

plt.plot(t,yout)
plt.title("Step Response of Second Order System")
plt.xlabel("Time (Seconds)")
plt.ylabel("Response")
plt.grid()
plt.show()
```
