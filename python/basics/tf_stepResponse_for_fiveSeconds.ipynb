# Code to Create Transfer Function of Second order sysetm and Plot Step Response

```python
import numpy as np 
import control as ct 
from control.matlab import *
import matplotlib.pyplot as plt
```

# Define Second Order Transfer functin using num and dem

```python
num=np.array([1,2])
den=np.array([1,2,5])
sys=ct.tf(num,den)
print(sys)
```

# Print Step response Info

```python
a=ct.step_info(sys)
print("Peak : ",a['Peak'])
print(a)
```

# Step Response (Define time period of step response form 0 to 5 second with dt=0.1

```python
#Define time
dt=0.1
sim_time=5
t = np.arange(0, sim_time, dt) 

#Step response
t,yout=ct.step_response(sys,t)

#Plot Graphs
plt.plot(t,yout)
plt.title("Step Response of Second Order System")
plt.xlabel("Time (Seconds)")
plt.ylabel("Response")
plt.grid()
plt.show()
```
